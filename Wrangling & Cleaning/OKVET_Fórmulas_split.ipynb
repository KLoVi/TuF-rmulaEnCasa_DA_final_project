```python
df = pd.read_excel(url)
```

```python
df.drop( columns = ["Medicamentos 2"] , inplace = True)
```

```python
patron = 'Nombre: (.*?)Presentación'
nombres_medicamentos = df['Medicamentos 1'].str.extractall(patron)
```

```python
nombres_medicamentos
```

```python
nombres_medicamentos.reset_index(inplace=True)
```

```python
nombres_medicamentos
```

```python
nombres_medicamentos.columns = ['indice_original', 'medicamento_num', 'nombre_medicamento']
```

```python
nombres_medicamentos
```

```python
nombres_medicamentos['medicamento_num'] += 1
```

```python
nombres_medicamentos
```

```python
# Aseguramos que 'indice_original' es del mismo tipo en ambos DataFrames para el merge
df.reset_index(inplace=True)
nombres_medicamentos['indice_original'] = nombres_medicamentos['indice_original'].astype(df.index.dtype)
```

```python
df.head(3)
```

```python
nombres_medicamentos.head(3)
```

```python
# Hacemos un merge entre el DataFrame original y los medicamentos extraídos
df_final = pd.merge(df, nombres_medicamentos, left_on='index', right_on='indice_original')
```
